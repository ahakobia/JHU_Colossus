# Merging Kalyn Data

```python
medal_count_df = pd.read_csv('../data_kalyn/medal_count.csv')
# Standardize column names to lowercase
medal_count_df.columns = medal_count_df.columns.str.lower()
```

```python
gdp_df = pd.read_csv('../data_kalyn/annual_GDP.csv')
gdp_df.columns = gdp_df.columns.str.lower()

count_ath_cnt_df = pd.read_csv('../data_kalyn/country_athlete_count.csv')
count_ath_cnt_df.columns = count_ath_cnt_df.columns.str.lower()

hdi_df = pd.read_csv('../data_kalyn/country_hdi.csv')
hdi_df.columns = hdi_df.columns.str.lower()

count_map_df = pd.read_csv('../data_kalyn/country_mapping.csv')
count_map_df.columns = count_map_df.columns.str.lower()

gender_index_df = pd.read_csv('../data_kalyn/gender_index.csv')
gender_index_df.columns = gender_index_df.columns.str.lower()

medalstats_df = pd.read_csv('../data_kalyn/medal_stats.csv')
medalstats_df.columns = medalstats_df.columns.str.lower()
medalstats_df = medalstats_df.rename(columns={'total': 'total_medals%'})
medalstats_df = medalstats_df.rename(columns={'participants': 'participants%'})
medalstats_df = medalstats_df.drop(columns=['host'])

# Merge the datasets step by step
merged_df = pd.merge(medal_count_df, gdp_df, on=['year', 'code'], how='left')
merged_df2 = pd.merge(merged_df, count_ath_cnt_df, on=['edition_id', 'year','code'], how='left')
merged_df3 = pd.merge(merged_df2, hdi_df, on=['code'], how='left')
merged_df4 = pd.merge(merged_df3, count_map_df, on=['code'], how='left')
merged_df5 = pd.merge(merged_df4, gender_index_df, on=['code'], how='left')
merged_df6 = pd.merge(merged_df5, medalstats_df, on=['year', 'code'], how='left')

merged_df6 = merged_df6.rename(columns={'total': 'total_medals'})

# Display the first few rows of the final merged dataset
merged_df6.head()
```

```python
final_path = 'merged_df_kalyn.csv'
merged_df6.to_csv(final_path, index=False)
```

# Kalyn + Jon + Arian Data

```python
merged_df_kalyn_Jon = pd.read_csv('merged_df_kalyn_Jon.csv')
merged_df_kalyn_Jon.columns = merged_df_kalyn_Jon.columns.str.lower()
merged_df_kalyn_Jon = merged_df_kalyn_Jon.drop(columns=['entity'])
merged_df_kalyn_Jon.tail()
```

```python
popcity_df = pd.read_csv('../Population-Arian/Population in the capital city,urban and rural areas.csv')
popcity_df.columns = popcity_df.columns.str.lower()

# Select only the 'series' and 'value' columns from popcity_df and add a suffix
popcity_df_filtered_0 = popcity_df[['country', 'year', 'series', 'value']].copy()
popcity_df_filtered_0 = popcity_df_filtered_0.rename(columns={'series': 'series_popcity', 'value': 'value_popcity'})

# Remove commas from 'value_popcity' and convert to float, then to int
popcity_df_filtered_0['value_popcity'] = popcity_df_filtered_0['value_popcity'].str.replace(',', '').astype(float).astype(int)

# Pivot the dataframe so that 'series_popcity' becomes columns and 'value_popcity' are the values
popcity_df_filtered = popcity_df_filtered_0.pivot_table(index=['country', 'year'], columns='series_popcity', values='value_popcity', aggfunc='first').reset_index()
```

```python
# Ensure country and year columns are correctly formatted
merged_df_kalyn_Jon['country'] = merged_df_kalyn_Jon['country'].str.strip().str.upper()
popcity_df_filtered['country'] = popcity_df_filtered['country'].str.strip().str.upper()

# Define a function to find the closest year within 5 years
def find_closest_year_within_5_years(row, df):
    country = row['country']
    year = row['year']
    available_years = df[df['country'] == country]['year']
    
    # Filter years within 5 years of the current year
    valid_years = available_years[(available_years >= year - 5) & (available_years <= year + 5)]
    
    # Find the closest valid year
    closest_year = valid_years.iloc[(valid_years - year).abs().argsort()[:1]]
    
    return closest_year.values[0] if not closest_year.empty else None

# Apply the function to find the closest year within 5 years for each row
merged_df_kalyn_Jon['closest_year'] = merged_df_kalyn_Jon.apply(find_closest_year_within_5_years, axis=1, df=popcity_df_filtered)

# Merge on country and closest year
merged_df_kalyn_Jon_Arian = pd.merge(
    merged_df_kalyn_Jon, 
    popcity_df_filtered, 
    left_on=['country', 'closest_year'], 
    right_on=['country', 'year'], 
    how='left'
)

# Drop the 'closest_year' and 'year_y' columns
merged_df_kalyn_Jon_Arian = merged_df_kalyn_Jon_Arian.drop(columns=['closest_year', 'year_y'])

# Rename 'year_x' back to 'year'
merged_df_kalyn_Jon_Arian = merged_df_kalyn_Jon_Arian.rename(columns={'year_x': 'year'})

# Display the first few rows of the final merged dataframe
merged_df_kalyn_Jon_Arian.head()
```

```python
popmigrants_df = pd.read_csv('../Population-Arian/Population International migrants and refugees.csv')
popmigrants_df.columns = popmigrants_df.columns.str.lower()

# Select only the 'series' and 'value' columns from popcity_df and add a suffix
popmigrants_df_filtered_0 = popmigrants_df[['country', 'year', 'series', 'value']].copy()
popmigrants_df_filtered_0 = popmigrants_df_filtered_0.rename(columns={'series': 'series_popmigrant', 'value': 'value_popmigrant'})

# Remove commas from 'value_popcity' and convert to float, then to int
popmigrants_df_filtered_0['value_popmigrant'] = popmigrants_df_filtered_0['value_popmigrant'].str.replace(',', '').astype(float).astype(int)

# Pivot the dataframe so that 'series_popcity' becomes columns and 'value_popcity' are the values
popmigrants_df_filtered = popmigrants_df_filtered_0.pivot_table(index=['country', 'year'], columns='series_popmigrant', values='value_popmigrant', aggfunc='first').reset_index()
```

```python
# Ensure country and year columns are correctly formatted
merged_df_kalyn_Jon_Arian['country'] = merged_df_kalyn_Jon_Arian['country'].str.strip().str.upper()
popmigrants_df_filtered['country'] = popmigrants_df_filtered['country'].str.strip().str.upper()

# Apply the function to find the closest year within 5 years for each row
merged_df_kalyn_Jon_Arian['closest_year'] = merged_df_kalyn_Jon_Arian.apply(find_closest_year_within_5_years, axis=1, df=popmigrants_df_filtered)

# Merge on country and closest year
merged_df_kalyn_Jon_Arian2 = pd.merge(
    merged_df_kalyn_Jon_Arian, 
    popmigrants_df_filtered, 
    left_on=['country', 'closest_year'], 
    right_on=['country', 'year'], 
    how='left'
)

# Drop the 'closest_year' and 'year_y' columns
merged_df_kalyn_Jon_Arian2 = merged_df_kalyn_Jon_Arian2.drop(columns=['closest_year', 'year_y'])

# Rename 'year_x' back to 'year'
merged_df_kalyn_Jon_Arian2 = merged_df_kalyn_Jon_Arian2.rename(columns={'year_x': 'year'})

# Display the first few rows of the final merged dataframe
merged_df_kalyn_Jon_Arian2.head()
```

```python
popdensity_df = pd.read_csv('../Population-Arian/Population surface area and density.csv')
popdensity_df.columns = popdensity_df.columns.str.lower()

# Select only the 'series' and 'value' columns from popcity_df and add a suffix
popdensity_df_filtered_0 = popdensity_df[['country', 'year', 'series', 'value']].copy()
popdensity_df_filtered_0 = popdensity_df_filtered_0.rename(columns={'series': 'series_popdensity', 'value': 'value_popdensity'})

# Remove commas from 'value_popcity' and convert to float, then to int
popdensity_df_filtered_0['value_popdensity'] = popdensity_df_filtered_0['value_popdensity'].str.replace(',', '').astype(float).astype(int)

# Pivot the dataframe so that 'series_popcity' becomes columns and 'value_popcity' are the values
popdensity_df_filtered = popdensity_df_filtered_0.pivot_table(index=['country', 'year'], columns='series_popdensity', values='value_popdensity', aggfunc='first').reset_index()
```

```python
# Ensure country and year columns are correctly formatted
merged_df_kalyn_Jon_Arian2['country'] = merged_df_kalyn_Jon_Arian2['country'].str.strip().str.upper()
popdensity_df_filtered['country'] = popdensity_df_filtered['country'].str.strip().str.upper()

# Apply the function to find the closest year within 5 years for each row
merged_df_kalyn_Jon_Arian2['closest_year'] = merged_df_kalyn_Jon_Arian2.apply(find_closest_year_within_5_years, axis=1, df=popdensity_df_filtered)

# Merge on country and closest year
merged_df_kalyn_Jon_Arian3 = pd.merge(
    merged_df_kalyn_Jon_Arian2, 
    popdensity_df_filtered, 
    left_on=['country', 'closest_year'], 
    right_on=['country', 'year'], 
    how='left'
)

# Drop the 'closest_year' and 'year_y' columns
merged_df_kalyn_Jon_Arian3 = merged_df_kalyn_Jon_Arian3.drop(columns=['closest_year', 'year_y'])

# Rename 'year_x' back to 'year'
merged_df_kalyn_Jon_Arian3 = merged_df_kalyn_Jon_Arian3.rename(columns={'year_x': 'year'})

# Display the first few rows of the final merged dataframe
merged_df_kalyn_Jon_Arian3.head()
```

```python
# Load the WHO data
who_df = pd.read_csv('../Population-Arian/WHO-COVID-19-global-data.csv')

# Parse the 'Date_reported' column as datetime and extract the year
who_df['Date_reported'] = pd.to_datetime(who_df['Date_reported'])
who_df['Year'] = who_df['Date_reported'].dt.year

# Sort the dataframe by 'Date_reported' to ensure the last date is at the end
who_df = who_df.sort_values(by=['Country_code', 'Year', 'Date_reported'])

# Group by 'Country_code', 'Country', and 'Year' and select the last entry for each group
last_reported_df = who_df.groupby(['Country_code', 'Country', 'Year']).last().reset_index()

# Keep only the relevant columns
who_df2 = last_reported_df[['Country', 'Year', 'Cumulative_cases', 'Cumulative_deaths']]

# Add '_covid' suffix to the WHO columns, except 'country' and 'year'
who_df2 = who_df2.rename(columns=lambda x: x + '_covid' if x not in ['Country', 'Year'] else x)

# Ensure 'year' is in the correct format and convert country names to uppercase
who_df2 = who_df2.rename(columns={'Year': 'year', 'Country': 'country'})
who_df2['country'] = who_df2['country'].str.upper()

# Convert the country names in the merged_df_kalyn_Jon_Arian3 to uppercase
merged_df_kalyn_Jon_Arian3['country'] = merged_df_kalyn_Jon_Arian3['country'].str.upper()

# Merge the WHO data with merged_df_kalyn_Jon_Arian3 on 'country' and 'year'
merged_df_kalyn_Jon_Arian3_final = pd.merge(
    merged_df_kalyn_Jon_Arian3, 
    who_df2, 
    on=['country', 'year'], 
    how='left'
)

# Display the first few rows of the final merged dataframe
merged_df_kalyn_Jon_Arian3_final.tail()
```

```python
final_path = 'merged_df_kalyn_Jon_Arian3_final.csv'
merged_df_kalyn_Jon_Arian3_final.to_csv(final_path, index=False)
```

# Kalyn + Jon + Arian + Adeh Data

```python
gdp_data_path = '../health_data_adeh/cleaning_process/cleandata/cleaned_gdp_imf.csv'
combined_data_path = '../health_data_adeh/cleaning_process/cleandata/combined_country_territory_data.csv'
combined_sporting_path = '../health_data_adeh/cleaning_process/cleandata/combined_sport_spending_data.csv'
mortality_data_path = '../health_data_adeh/cleaning_process/cleandata/processed_mortality_data.csv'
oecd_data_path = '../health_data_adeh/cleaning_process/cleandata/processed_oecd_health_govspend_data.csv'
public_finances_path = '../health_data_adeh/cleaning_process/cleandata/public_finances_clean.xlsx'

merged_df_kalyn_Jon_Arian3_final = pd.read_csv('merged_df_kalyn_Jon_Arian3_final.csv')

combined_data = pd.read_csv(combined_data_path)
gdp_data = pd.read_csv(gdp_data_path)
mortality_data = pd.read_csv(mortality_data_path)
oecd_data = pd.read_csv(oecd_data_path)
combined_sporting_data = pd.read_csv(combined_sporting_path)
public_finances_data = pd.read_excel(public_finances_path)
```

```python
# Standardize the column names to lowercase
combined_data.columns = combined_data.columns.str.lower()
gdp_data.columns = gdp_data.columns.str.lower()
mortality_data.columns = mortality_data.columns.str.lower()
oecd_data.columns = oecd_data.columns.str.lower()
combined_sporting_data.columns = combined_sporting_data.columns.str.lower()
public_finances_data.columns = public_finances_data.columns.str.lower()

# Ensure country values are fully uppercase for merging
combined_data['country'] = combined_data['country'].str.upper()
gdp_data['country'] = gdp_data['country'].str.upper()
mortality_data['country'] = mortality_data['country'].str.upper()
oecd_data['country'] = oecd_data['country'].str.upper()
combined_sporting_data['country'] = combined_sporting_data['country'].str.upper()
public_finances_data['country'] = public_finances_data['country'].str.upper()

# Replace spaces in combined_sporting_data column names with underscores
combined_sporting_data.columns = combined_sporting_data.columns.str.replace(' ', '_')

# Rename GDP column in gdp_data
gdp_data = gdp_data.rename(columns={'gdp': 'gdp%change'})

# Merge with the main dataframe
merged_df = merged_df_kalyn_Jon_Arian3_final.copy()

# Merge public finances data
merged_df = pd.merge(merged_df, public_finances_data.drop(columns=['ifscode']), 
                     on=['country', 'year'], how='left')

# Merge combined data
merged_df = pd.merge(merged_df, combined_data, 
                     on=['country', 'year'], how='left')

# Merge GDP data
merged_df = pd.merge(merged_df, gdp_data, 
                     on=['country', 'year'], how='left')

# Merge mortality data (using 'code' and 'year' for the merge)
merged_df = pd.merge(merged_df, mortality_data.drop(columns=['country']), 
                     on=['code', 'year'], how='left')

# Merge OECD data
merged_df = pd.merge(merged_df, oecd_data, 
                     on=['country', 'year'], how='left')

# Merge combined sporting data
merged_df = pd.merge(merged_df, combined_sporting_data, 
                     on=['country', 'year'], how='left')
merged_df = merged_df.drop(columns=['countries','region'])
# Display the first few rows of the final merged dataframe
merged_df.tail()
```

```python
final_path = 'FINAL_DATA.csv'
merged_df.to_csv(final_path, index=False)
```
