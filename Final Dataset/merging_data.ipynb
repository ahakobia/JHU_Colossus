```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
medal_count_df = pd.read_csv('../data_kalyn/medal_count.csv')
# Standardize column names to lowercase
medal_count_df.columns = medal_count_df.columns.str.lower()
```

```python
gdp_df = pd.read_csv('../data_kalyn/annual_GDP.csv')
gdp_df.columns = gdp_df.columns.str.lower()

count_ath_cnt_df = pd.read_csv('../data_kalyn/country_athlete_count.csv')
count_ath_cnt_df.columns = count_ath_cnt_df.columns.str.lower()

hdi_df = pd.read_csv('../data_kalyn/country_hdi.csv')
hdi_df.columns = hdi_df.columns.str.lower()

count_map_df = pd.read_csv('../data_kalyn/country_mapping.csv')
count_map_df.columns = count_map_df.columns.str.lower()

gender_index_df = pd.read_csv('../data_kalyn/gender_index.csv')
gender_index_df.columns = gender_index_df.columns.str.lower()

# Merge the datasets step by step
merged_df = pd.merge(medal_count_df, gdp_df, on=['year', 'code'], how='left')
merged_df2 = pd.merge(merged_df, count_ath_cnt_df, on=['edition_id', 'code'], how='left')
merged_df3 = pd.merge(merged_df2, hdi_df, on=['code'], how='left')
merged_df4 = pd.merge(merged_df3, count_map_df, on=['code'], how='left')
merged_df5 = pd.merge(merged_df4, gender_index_df, on=['code'], how='left')

# Display the first few rows of the final merged dataset
merged_df5.head()
```

```python
final_path = 'merged_df_kalyn.csv'
merged_df5.to_csv(final_path, index=False)
```

```python
# Load the datasets
merged_all_data = pd.read_csv('health_data_adeh/merged_all_data.csv', low_memory=False)
country_mapping = pd.read_csv('data_kalyn/country_mapping.csv')
country_hdi = pd.read_csv('data_kalyn/country_hdi.csv')
medal_count = pd.read_csv('data_kalyn/medal_count.csv')
```

```python
# Standardize column names
merged_all_data.rename(columns=lambda x: x.strip().lower(), inplace=True)
country_mapping.rename(columns=lambda x: x.strip().lower(), inplace=True)
country_hdi.rename(columns=lambda x: x.strip().lower(), inplace=True)
medal_count.rename(columns=lambda x: x.strip().lower(), inplace=True)

# Ensure 'code' and 'year' are strings for consistency
merged_all_data['code'] = merged_all_data['country code'].astype(str)
merged_all_data['year'] = merged_all_data['year'].astype(str)
country_mapping['code'] = country_mapping['code'].astype(str)
country_hdi['code'] = country_hdi['code'].astype(str)
medal_count['code'] = medal_count['code'].astype(str)
medal_count['year'] = medal_count['year'].astype(str)
```

```python
# Merge datasets
medal_count = medal_count.merge(country_mapping, on='code', how='left')
medal_count = medal_count.merge(country_hdi, on='code', how='left')
medal_count = medal_count.merge(merged_all_data, on=['code', 'year'], how='left')
```
