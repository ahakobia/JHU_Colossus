```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Merging Kalyn Data

```python
medal_count_df = pd.read_csv('../data_kalyn/medal_count.csv')
# Standardize column names to lowercase
medal_count_df.columns = medal_count_df.columns.str.lower()
```

```python
gdp_df = pd.read_csv('../data_kalyn/annual_GDP.csv')
gdp_df.columns = gdp_df.columns.str.lower()

count_ath_cnt_df = pd.read_csv('../data_kalyn/country_athlete_count.csv')
count_ath_cnt_df.columns = count_ath_cnt_df.columns.str.lower()

hdi_df = pd.read_csv('../data_kalyn/country_hdi.csv')
hdi_df.columns = hdi_df.columns.str.lower()

count_map_df = pd.read_csv('../data_kalyn/country_mapping.csv')
count_map_df.columns = count_map_df.columns.str.lower()

gender_index_df = pd.read_csv('../data_kalyn/gender_index.csv')
gender_index_df.columns = gender_index_df.columns.str.lower()

# Merge the datasets step by step
merged_df = pd.merge(medal_count_df, gdp_df, on=['year', 'code'], how='left')
merged_df2 = pd.merge(merged_df, count_ath_cnt_df, on=['edition_id', 'code'], how='left')
merged_df3 = pd.merge(merged_df2, hdi_df, on=['code'], how='left')
merged_df4 = pd.merge(merged_df3, count_map_df, on=['code'], how='left')
merged_df5 = pd.merge(merged_df4, gender_index_df, on=['code'], how='left')

# Display the first few rows of the final merged dataset
merged_df5.head()
```

```python
final_path = 'merged_df_kalyn.csv'
merged_df5.to_csv(final_path, index=False)
```

# Kalyn + Jon Data

```python
merged_df_kalyn = pd.read_csv('merged_df_kalyn.csv')
merged_df_kalyn.columns = merged_df_kalyn.columns.str.lower()
```

```python
human_freedom_df = pd.read_csv('../Political_Data_Jon/2023-Human-Freedom-Index-Data.csv')
human_freedom_df.columns = human_freedom_df.columns.str.lower()
# Rename 'iso' to 'code'
human_freedom_df = human_freedom_df.rename(columns={'iso': 'code'})
merged_df_k_J = pd.merge(merged_df_kalyn, human_freedom_df, on=['year', 'code'], how='left')

# Sort the dataframe by 'year' and 'code'
merged_df_k_J = merged_df_k_J.sort_values(by=['year', 'code'])

# Filter the dataframe to include only rows where 'year' is less than 2024
merged_df_k_J2 = merged_df_k_J[merged_df_k_J['year'] < 2024]

# Display the first few rows
merged_df_k_J2.tail()
```

```python
final_path = 'merged_df_kalyn_Jon.csv'
merged_df_k_J3.to_csv(final_path, index=False)
```

# Kalyn + Jon + Arian Data

```python
merged_df_kalyn_Jon = pd.read_csv('merged_df_kalyn_Jon.csv')
merged_df_kalyn_Jon.columns = merged_df_kalyn_Jon.columns.str.lower()
merged_df_kalyn_Jon = merged_df_kalyn_Jon.drop(columns=['entity'])
merged_df_kalyn_Jon.tail()
```

```python
popcity_df = pd.read_csv('../Population-Arian/Population in the capital city,urban and rural areas.csv')
popcity_df.columns = popcity_df.columns.str.lower()

# Select only the 'series' and 'value' columns from popcity_df and add a suffix
popcity_df_filtered = popcity_df[['country', 'year', 'series', 'value']].copy()
popcity_df_filtered = popcity_df_filtered.rename(columns={'series': 'series_popcity', 'value': 'value_popcity'})
```

```python
# Ensure country and year columns are correctly formatted
merged_df_kalyn_Jon['country'] = merged_df_kalyn_Jon['country'].str.strip().str.upper()
popcity_df_filtered['country'] = popcity_df_filtered['country'].str.strip().str.upper()

# Define a function to find the closest year within 5 years
def find_closest_year_within_5_years(row, popcity_df_filtered):
    country = row['country']
    year = row['year']
    available_years = popcity_df_filtered[popcity_df_filtered['country'] == country]['year']
    
    # Filter years within 5 years of the current year
    valid_years = available_years[(available_years >= year - 5) & (available_years <= year + 5)]
    
    # Find the closest valid year
    closest_year = valid_years.iloc[(valid_years - year).abs().argsort()[:1]]
    
    return closest_year.values[0] if not closest_year.empty else None

# Apply the function to find the closest year within 5 years for each row
merged_df_kalyn_Jon['closest_year'] = merged_df_kalyn_Jon.apply(find_closest_year_within_5_years, axis=1, popcity_df_filtered=popcity_df_filtered)

# Merge on country and closest year
merged_df_kalyn_Jon_Arian = pd.merge(
    merged_df_kalyn_Jon, 
    popcity_df_filtered, 
    left_on=['country', 'closest_year'], 
    right_on=['country', 'year'], 
    how='left'
)

# Drop the 'closest_year' and 'year_y' columns
merged_df_kalyn_Jon_Arian = merged_df_kalyn_Jon_Arian.drop(columns=['closest_year', 'year_y'])

# Rename 'year_x' back to 'year'
merged_df_kalyn_Jon_Arian = merged_df_kalyn_Jon_Arian.rename(columns={'year_x': 'year'})

# Display the first few rows of the final merged dataframe
merged_df_kalyn_Jon_Arian.head()
```
